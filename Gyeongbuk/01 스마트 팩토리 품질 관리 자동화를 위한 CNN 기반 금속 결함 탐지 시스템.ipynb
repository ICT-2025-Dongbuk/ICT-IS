## 2. 데이터 로드 및 전처리

```python
# 2.1. 데이터 경로 설정
my_data_dir = './casting_data/casting_data/'
train_path = my_data_dir + 'train/'
test_path = my_data_dir + 'test/'

# 2.2. 이미지 크기 확인
# 데이터셋의 첫 번째 이미지의 크기를 확인하여 모델 입력 크기를 결정합니다.
image = Image.open(glob.glob(train_path + '/*/*.*')[0])
width, height = image.size
print(f"이미지 크기: {width} x {height}")
```

```python
# 데이터 분포 확인
# 학습 및 테스트 데이터셋의 정상(ok_front) 및 불량(def_front) 이미지 개수를 확인합니다.
def count_images(path, data_type):
    """지정된 경로의 이미지 개수를 세어 데이터프레임으로 반환합니다."""
    ok_path = os.path.join(path, 'ok_front')
    def_path = os.path.join(path, 'def_front')

    ok_count = len(os.listdir(ok_path))
    def_count = len(os.listdir(def_path))

    df = pd.DataFrame({'Category': ['OK', 'Defect'], 'Count': [ok_count, def_count]})
    df['Type'] = data_type
    return df

train_counts = count_images(train_path, 'Train')
test_counts = count_images(test_path, 'Test')

data_dist = pd.concat([train_counts, test_counts])

# 데이터 분포 시각화
px.bar(data_dist, x='Category', y='Count', color='Type', barmode='group',
       title='Train vs. Test Data Distribution',
       labels={'Count': '이미지 수', 'Category': '분류'})
```

```python
# 이미지 샘플 시각화
# 정상 및 불량 이미지 샘플을 각각 5개씩 시각화하여 확인합니다.

def plot_image_samples(path, category, title, n_samples=5):
    """지정된 카테고리의 이미지 샘플을 시각화합니다."""
    fig, axes = plt.subplots(1, n_samples, figsize=(15, 3))
    fig.suptitle(title, fontsize=16)

    img_files = os.listdir(os.path.join(path, category))[:n_samples]

    for i, img_file in enumerate(img_files):
        img_path = os.path.join(path, category, img_file)
        img = plt.imread(img_path)
        axes[i].imshow(img)
        axes[i].axis('off')
    plt.show()

# 정상 이미지 샘플
plot_image_samples(train_path, 'ok_front', 'OK image sample')

# 불량 이미지 샘플
plot_image_samples(train_path, 'def_front', 'Defect image sample')
```

## 3. Data Augmentation

```python
# 데이터 증강 (Data Augmentation)
# ImageDataGenerator를 사용하여 이미지 회전, 이동, 확대/축소, 뒤집기 등 다양한 변형을 적용하여
# 학습 데이터의 다양성을 높이고, 모델이 과적합되는 것을 방지합니다.

# 이미지 데이터 제너레이터 생성
datagen = ImageDataGenerator(
    rescale=1./255,          # 픽셀 값을 0~1 사이로 정규화
    rotation_range=20,       # 20도 내에서 이미지 회전
    width_shift_range=0.2,   # 20% 내에서 수평 이동
    height_shift_range=0.2,  # 20% 내에서 수직 이동
    shear_range=0.2,         # 20% 내에서 전단 변환
    zoom_range=0.2,          # 20% 내에서 확대/축소
    horizontal_flip=True,    # 수평 뒤집기
    fill_mode='nearest',     # 변형 시 생기는 빈 공간을 가까운 픽셀로 채움
    validation_split=0.2     # 검증 데이터 분할 비율
)

# 학습 및 검증 데이터셋 생성
train_set = datagen.flow_from_directory(
    train_path,
    target_size=(300, 300),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

test_set = datagen.flow_from_directory(
    test_path,
    target_size=(300, 300),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

## 4. Modeling

```python
# CNN 모델 구축
# 합성곱 신경망(CNN) 모델을 Sequentia를 사용하여 구성합니다.

model = Sequential([
    # 첫 번째 합성곱 블록
    Conv2D(32, (3, 3), activation='relu', input_shape=(300, 300, 3)),
    MaxPooling2D(2, 2),

    # 두 번째 합성곱 블록
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),

    # 세 번째 합성곱 블록
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),

    # 네 번째 합성곱 블록
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),

    # 완전 연결 레이어
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5), # 과적합 방지를 위한 드롭아웃
    Dense(1, activation='sigmoid') # 이진 분류를 위한 출력 레이어
])

model.summary()
```

```python
# 모델 컴파일 및 학습

# 모델 컴파일
# adam 옵티마이저와 binary_crossentropy 손실 함수를 사용하여 모델을 컴파일합니다.
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# 콜백 설정
# 조기 종료(EarlyStopping)와 모델 체크포인트(ModelCheckpoint) 콜백을 설정하여
# 학습 중 검증 손실이 개선되지 않으면 학습을 중단하고, 가장 좋은 모델을 저장합니다.
early_stop = EarlyStopping(monitor='val_loss', mode='min', patience=5, restore_best_weights=True)
checkpoint = ModelCheckpoint('casting_product_detection.keras', monitor='val_loss', mode='min', save_best_only=True)

# 모델 학습
# fit_generator를 사용하여 증강된 데이터로 모델을 학습시킵니다.
# fit을 사용하여 증강된 데이터로 모델을 학습시킵니다.
n_epochs = 1
results = model.fit(
    train_set,
    epochs=n_epochs,
    validation_data=test_set,
    callbacks=[early_stop, checkpoint]
)

# 학습 이력 저장
model_history = results.history
json.dump(model_history, open('model_history.json', 'w'))
```

## 5. Model Evaluation

```python
# 모델 평가

# 학습 결과 데이터프레임 확인
# 학습 과정에서의 손실 및 정확도를 데이터프레임으로 확인합니다.
losses = pd.DataFrame(model_history)
losses.index = range(1, len(losses) + 1)
print(losses.head(3))
```

```python
# 손실 및 정확도 그래프
# HoloViews를 사용하여 학습 및 검증 데이터에 대한 손실과 정확도 변화를 시각화합니다.
g = (
    hv.Curve(losses.loss, label='Training Loss') *
    hv.Curve(losses.val_loss, label='Validation Loss') *
    hv.Curve(losses.accuracy, label='Training Accuracy') *
    hv.Curve(losses.val_accuracy, label='Validation Accuracy')
)

g.opts(
    opts.Curve(xlabel="Epochs", ylabel="Loss / Accuracy", width=700, height=400,
               tools=['hover'], show_grid=True, title='Model Evaluation')
).opts(legend_position='bottom')
```

Since the loss at the time of training and validation are steadily decreasing for each epoch, and the accuracy at the time of training and the validation are steadily increasing, it can be said that the training is generally successful.

```python
# 혼동 행렬 및 분류 리포트
# 테스트 데이터에 대한 예측을 수행하고, 혼동 행렬과 분류 리포트를 통해 모델의 성능을 평가합니다.

# 예측 수행
pred_probability = model.predict(test_set)
predictions = pred_probability > 0.5

# 혼동 행렬 시각화
plt.figure(figsize=(10, 6))
plt.title("Confusion Matrix", size=20, weight='bold')
sns.heatmap(
    confusion_matrix(test_set.classes, predictions),
    annot=True,
    annot_kws={'size': 14, 'weight': 'bold'},
    fmt='d',
    xticklabels=['Defect', 'OK'],
    yticklabels=['Defect', 'OK']
)
plt.tick_params(axis='both', labelsize=14)
plt.ylabel('Actual', size=14, weight='bold')
plt.xlabel('Predicted', size=14, weight='bold')
plt.show()

# 분류 리포트 출력
print(classification_report(test_set.classes, predictions, digits=3))
```
